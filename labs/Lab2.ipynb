```python
import random
from random import randint
import copy
import numpy as num
import time

def create_array():
    a = [random.randint(1, 50)for i in range(10)]
    return a

base_array = create_array()
print('Initial array: ', base_array)
```

```python
#Бинарный поиск
def binary_search(array, element, start, end):
    if start > end:
        return -1

    mid = (start + end) // 2
    if element == array[mid]:
        return mid

    if element < array[mid]:
        return binary_search(array, element, start, mid-1)
    else:
        return binary_search(array, element, mid+1, end) 

array = copy.deepcopy(base_array)

add_el = int(input('Введите элемент для вставки: '))
index = int(input('Введите индекс места, на которое нужно вставить элемент: '))
array.insert(index, add_el)
print('Array with added element: ', array, '\n')

delete_index = int(input('Введите индекс элемента, который нужно удалить: '))
array.pop(delete_index)
print('Array with deleted element: ', array, '\n')

array.sort()
print('Sorted array: ', array, '\n')

start_time = time.perf_counter()

element = int(input('Введите элемент для поиска: '))
print('Index of element {}: {}'.format(element, binary_search(array, element, 0, len(array))))

print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))
```

```python
#Бинарное дерево
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def inorder(root): #отсортированный (inorder) обход
    if root is not None:
        inorder(root.left) #обход левого поддерева
        print(str(root.key) + "->", end=' ')  #обход корня
        inorder(root.right) #обходим правого поддерева

def insert(node, key):
    if node is None:
        return TreeNode(key) #возвращаем новый узел, если дерево пустое

    if key < node.key: #идем в нужное место и вставляет узел
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)
    return node

def minValueNode(node): #поиск inorder-преемника
    current = node
    #найдем крайний левый лист — он и будет inorder-преемником
    while(current.left is not None):
        current = current.left
    return current

def find(root, key): #сравниваем значение с узлами
    if key < root.key:
        root.left = find(root.left, key)
    elif key > root.key:
        root.right = find(root.right, key)
    else:
        print(str(root.key) + ' is found!')
            
    return root
            
def deleteNode(root, key):
    if root is None:
        return root #возвращаем, если дерево пустое

    if key < root.key:
        root.left = deleteNode(root.left, key)
    elif(key > root.key):
        root.right = deleteNode(root.right, key)
    else:
        #если у узла только один дочерний узел или вообще их нет        
        if root.left is None:
            temp = root.right
            root = None
            return temp

        elif root.right is None:
            temp = root.left
            root = None
            return temp

        #усли у узла два дочерних узла, помещаем центрированного преемника
        #на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.key = temp.key
        #удаляем inorder-преемниа
        root.right = deleteNode(root.right, temp.key)
        
    return root

root = None
root = insert(root, 8)
root = insert(root, 3)
root = insert(root, 1)
root = insert(root, 6)
root = insert(root, 7)
root = insert(root, 10)
root = insert(root, 14)
root = insert(root, 4)

print('In-order bypass: ', end = ' ')
inorder(root)

start_time = time.perf_counter()

print('\n After finding 7 ')
root = find(root, 7)

print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

print('\n After deleting 14 ')
root = deleteNode(root, 14)

print('In-order bypass with deleted node: ', end = ' ')
inorder(root)
```

```python
#Метод Фибоначчи
def fibonacci(array, element):
    array.sort()
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(array)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1;
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(array)-1))
        if (array[i] < element):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (array[i] > element):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else :
            return i
    if(fibM_minus_1 and index < (len(array)-1) and array[index+1] == element):
        return index+1;
    return -1

array = copy.deepcopy(base_array)

start_time = time.perf_counter() 
        
n = int(input('Введите элемент для поиска: '))
print('Значение элемента {}: {}'.format(n, fibonacci(array, n)), '\n') 

print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))

add_el = int(input('Введите элемент для вставки: '))
index = int(input('Введите индекс места, на которое нужно вставить элемент: '))
array.insert(index, add_el)
print('Array with added element: ', array, '\n')

delete_index = int(input('Введите индекс элемента, который нужно удалить: '))
array.pop(delete_index)
print('Array with deleted element: ', array)
```

```python
#Интерполяционный
def interpolation_search(array, value):
    start = 0
    end = (len(array) - 1)
    while start <= end and value >= array[start] and value <= array[end]:
        #вычисление вероятной позиции искомого элемента
        index = start + int(((float(end - start)/(array[end] - array[start]))*(value - array[start])))
        if array[index] == value:
            return index
        if array[index] < value: #пересчет индекса и поиск в правой половине
            start = index + 1;
        else:
            end = index - 1; #пересчет индекса и поиск в левой половине
    return -1

array = copy.deepcopy(base_array)

add_el = int(input('Введите элемент для вставки: '))
index = int(input('Введите индекс места, на которое нужно вставить элемент: '))
array.insert(index, add_el)
print('Array with added element: ', array, '\n')

delete_index = int(input('Введите индекс элемента, который нужно удалить: '))
array.pop(delete_index)
print('Array with deleted element: ', array, '\n')

array.sort()
print('Sorted array: ', array, '\n')

start_time = time.perf_counter()

element = int(input('Введите элемент для поиска: '))
print('Index of element {}: {}'.format(element, interpolation_search(array, element)))
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))
```

```python
#Метод цепочек
class chain_rehash():
    def __init__(self):
        self.rhash = [None] * 11
        
    def keys(self, element): 
        key = 0
        for i in range(len(element)):
            key = key + ord(element[i])
        return int(key % 11)

    def add(self, element):
        original_hash = self.keys(element)
        if self.rhash[original_hash] is None:
            self.rhash[original_hash] = [element]
            return
        else:
            length = len(self.rhash[original_hash])
            self.rhash[original_hash].append(element)
            return
    
    def search(self, element): 
        key = self.keys(element)
        if self.rhash[key] is not None:
            for i in range(len(self.rhash[key])):
                if self.rhash[key][i] == element:
                    return key, i
        return None, None
            
    def delete(self, element):
        key, i = self.search(element)
        if key is not None:
            del(self.rhash[key][i])
            print('Элемент {} удален'.format(key))
        else:
            print('Элемента {} нет в таблице'.format(key))
        
    def Print(self): 
        hash = 0
        for x in self.rhash:
            print('[hash: %i]' % hash, end = " ")
            if x is None:
                print('*empty*')
            else:
                for y in x:
                    print(y, end = " ")
                print(" ")
            hash += 1  
        
a = chain_rehash()
s = input('Хотите добавить новый идентификатор в таблицу? Введите да или нет ')
if s == 'да':
    while s == 'да':
        id = (input())
        a.add(id)
        print ('Хотите добавить новый идентификатор в таблицу? Введите да или нет ')
        s=(input())
        
a.Print()

print ('Хотите найти идентификатор в таблице? Введите да или нет ')
k=(input())
if k == 'да':
    while k == 'да':
        id=(input())
        hash = a.search(id)
        print(hash)
        print('Хотите найти идентификатор в таблице? Введите да или нет ')
        k = (input())
        
print('Хотите удалить идентификатор из таблицы? Введите да или нет')
s = (input())
if s == 'да':
    while s == 'да':
        id = (input())
        a.delete(id)
        print ('Хотите удалить идентификатор из таблицы? Введите да или нет')
        s = (input())
        
a.Print()
```
