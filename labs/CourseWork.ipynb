```python
# функция для проверки вхождения элемента в диапазон чисел
def check_range(n, a, b):
    if not a <= n <= b:
        return False
    return True

# функция для проверки равенства количества элементов массива заданному значению
def check_equal(n, arr):
    if not n == len(arr):
        return False
    return True

# функция для проверки латинских прописных и строчных букв
def check_ascii(arr):
    for i in range(len(arr)):
        if not (65<=ord(arr[i])<=90 or 97<=ord(arr[i])<=122):
            return False
    return True

# функция для проверки латинских строчных букв 'a' и 'b'
def check_ascii_limited(arr):
    for i in range(len(arr)):
        if not 97<=ord(arr[i])<=98:
            return False
    return True

# функция для проверки нахождения каждого элемента массива в диапазоне чисел
def check_arr(arr, a, b):
    for elem in arr:
        if not a <= elem <= b:
            return False
    return True

# функция для проверки является ли число вещественным и принадлежит диапазону [0,1]
def check_float(num):
    if abs(1-num) <= 1:
        return True
    return False

# функция для проверки расположения элементов массива в убывающем порядке
def check_decrease(arr):
    for i in range(len(arr)-1):
        if not arr[i] >= arr[i+1]:
            return False
    return True
```

Так сложилось, что в каждой из этих квартир находится ровно один покемон какого-то типа. Возможно, что в каких-то квартирах находятся покемоны одинаковых типов. Входные данные\

Во второй строке следует строка s длины n, состоящая из строчных и прописных букв английского алфавита, причем i-я буква равна типу покемона, который находится в квартире номер i.

Выходные данные\
Выведите минимальное число квартир, которые должен посетить Сергей Б., чтобы поймать покемонов всех типов, которые есть в доме.

Примечание\
В первом тестовом примере Сергей Б. может, например, начать с квартиры 1 и закончить в квартире 2.

Во втором тестовом примере Сергей Б. может, например, начать с квартиры 4 и закончить в квартире 6.

В третьем тестовом примере Сергей Б. должен начать с квартиры 2 и закончить в квартире 6.

```python
# функция для проверки нахождения элемента в массиве
def check(temp, el):
    for i in temp:
        if i == el:
            return 1

def flats(s):
    # запоминаем первый элемент массива, инициализируем точки входа и выхода
    temp, ent, out = [s[0]], 0, 0
    for i in range(len(s)-1):
        if s[i] == s[i+1] : # сравниваем соседние элементы
            if temp[0] == s[i+1]: # если встречался только данный элемент 
                ent = i+1 # смещаем точку входа
        elif check(temp, s[i+1]) != 1: # проверяем, находится элемент в массиве
            out = i+1 # если нет, то смещаем точку выхода
            temp.append(s[i+1]) # и добавляем в массив

    return out-ent+1 

try:
    n = int(input('Введите количество квартир: ')) 
    if not check_range(n, 1, 100000): sys.exit('Выход за пределы диапазона! 1 <= n <= 100000')
    s = list(map(str, input('Введите покемонов в квартирах: ')))
    if not check_ascii(s): sys.exit('Входные данные должны состоять из строчных и прописных букв английского алфавита!')
    if not check_equal(n, s): sys.exit('Количество введенных покемонов не соответствует номеру квартир!')
    else: print('Минимальное число квартир для посещения = ', flats(s))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code)
```

Вам дан массив целых чисел 𝑎1,𝑎2,…,𝑎𝑛. За одну операцию вы можете заменить один любой элемент массива на любое вещественное число. Определите минимальное число операций, необходимое, чтобы сделать массив хорошим.

Далее следуют описания 𝑡 наборов входных данных, по две строки на набор.

Первая строка содержит одно целое число 𝑛 (1≤𝑛≤70): количество чисел в массиве.

Вторая строка содержит 𝑛 целых чисел 𝑎1,𝑎2,…,𝑎𝑛 (−100≤𝑎𝑖≤100): изначальный массив.

В третьем примере массив хороший изначальное.

Во четвертом примере один из возможных хороших массивов — [−2.5⎯⎯⎯⎯⎯⎯⎯⎯,−2,−1.5⎯⎯⎯⎯⎯⎯⎯⎯,−1,−0.5⎯⎯⎯⎯⎯⎯⎯⎯,0].

```python
def func(n, arr):
    diff = []
    for i in range(1, n): # заносим в массив разность между соседними элементами
        diff.append(arr[i]-arr[i-1])
        
    # создаем словарь из массива с полученными разностями и считаем количество повторений
    frequency = {diff.count(x):x for x in set(diff)} 
    
    # если кол-во макс-ых разностей = длине массива
    if max(frequency.keys()) == len(diff): 
        # массив последовательный и менять элементы не нужно
        count = 0 
    # если разница длины массива и кол-ва макс-ых разностей = 1 
    elif len(diff) - max(frequency.keys()) == 1: 
        # счетчик равен количеству чисел с неповторяющейся разницей + 1
        count = len(diff) - max(frequency.keys()) + 1
    else: 
        # счетчик равен количеству чисел с неповторяющейся разницей
        count = len(diff) - max(frequency.keys()) 
    
    return count

try:
    t = int(input('Введите количество наборов входных данных: '))
    if not check_range(t, 1, 100): sys.exit('Выход за пределы диапазона! 1 <= t <= 100') 
    for i in range(t):
        n = int(input('Введите количество элементов массиве: '))
        if not check_range(n, 1, 70): sys.exit('Выход за пределы диапазона! 1 <= n <= 70')
        arr = list(map(int, input('Введите массив: ').split()))
        if not check_arr(arr, -100, 100): sys.exit('Выход за пределы диапазона! -100 <= a[i] <= 100')
        if not check_equal(n, arr): sys.exit('Количество элементов массива не соответствует заданному номеру!')
        elif len(arr) > 1: print('Минимальное количество элементов, которые нужно заменить = ', func(n, arr))
        else: print('Минимальное количество элементов, которые нужно заменить = 0')
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code)
```

либо перекрасить какой-то носок в любой цвет 𝑐′ (1≤𝑐′≤𝑛),

Два носка образуют пару, если это левый и правый носок одинакового цвета. Какую минимальное количество денег нужно заплатить Фениксу, чтобы собрать 𝑛/2 пар? Каждый носок должен попасть ровно в одну пару.

Входные данные\
Входные данные состоят из нескольких наборов. В первой строке задано одно целое число 𝑡 (1≤𝑡≤1000) — количество наборов входных данных.

В следующей строке заданы 𝑛 целых чисел 𝑐𝑖 (1≤𝑐𝑖≤𝑛) — цвета носков. Первые 𝑙 носков — левые, остальные 𝑟 носков — правые.

Гарантируется, что сумма 𝑛 по всем наборам входных данных не превосходит 2⋅105.

Выходные данные\
Для каждого набора входных данных, выведите одно число — минимальное количество денег, которое нужно заплатить Фениксу, чтобы собрать 𝑛/2 пар носков. Каждый носок должен попасть ровно в одну пару.

Пример\

Примечание\
В первом наборе, Феникс может заплатить 2 доллара для того, чтобы:

перекрасить носок 3 в цвет 2.\
В результате он получит 3 пары. Например, пары (1,4), (2,5) и (3,6).\
Во втором наборе, Феникс может заплатить 3 доллара, чтобы:

превратить носок 6 из правого в левый,\
перекрасить носок 3 в цвет 1,\
перекрасить носок 4 в цвет 1.\
В результате он получит 3 пары. Например, пары (1,3), (2,4) и (5,6).

```python
def change(one, two):
    while len(one) != len(two): # пока массивы не равны 
        one.append(two[0]) # добавляем в меньший элемент из большего
        del two[0] # и удаляем его из большего
    return one, two

def check(one, two):
    for temp in one[:]: # проходимся по образам массивов
        for elem in two[:]:
            if temp == elem: # если находим одинаковые
                two.remove(elem) # то удаляем из обоих массивов
                one.remove(temp)
                break
    return one, two  
    
def fill(s, c):
    # заносим цвета носков в 'левый' и 'правый' массивы
    count, j = 0, - 1
    l, r = [], []
    for i in range(int(s[1])):
        l.append(int(c[i]))
        j = i
    for i in range(int(s[2])):
        r.append(int(c[j+1])) 
        j += 1

    a, b = max(l, r), min(l, r) # определяем, макс и мин массив 
    count = (len(a)-len(b))//2 # считаем предварительный ответ
    change(b, a) # делаем 'левый' и 'правый' массивы равными
    check(a, b) # убираем носки одного цвета в разных массивах
    
    return count + len(l) # добавляем к счетчику оставшуюся длину одного массива

try:
    t = int(input("Введите количество наборов входных данных: "))
    if not check_range(t, 1, 1000): sys.exit('Выход за пределы диапазона! 1 <= t <= 1000') 
    for i in range(t):
        s = list(map(int, input("Введите общее количество носков, количество левых и правых носков: ").split()))
        if not check_range(s[0], 2, 2*pow(10,5)) or s[0]%2 != 0:
            sys.exit('Выход за пределы диапазона! n - четное число и 2 <= n <= 2*10^5')
        if not (check_range(s[1], 1, s[0]) or check_range(s[2], 1, s[0])) or s[1]+s[2]!=s[0]:
            sys.exit('Выход за пределы диапазона! 0 <= l,r <=n; l+r=n')
        c = list(map(int, input("Введите цвета носков: ").split()))
        if not check_arr(c, 1, s[0]): sys.exit('Выход за пределы диапазона! 1 <= c[i] <= n')
        if not check_equal(s[0], c): sys.exit('Количество элементов массива не соответствует заданному номеру!')
        print('Минимальное количество денег, которое нужно заплатить = ', fill(s, c))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def func(s, arr, rule):
    prob = 1.0 # задаем максимальную вероятность
    order = s[0] # переменной порядка присваиваем значение количества элементов
    
    # пока порядок < длины массива и элемент = своему индексу + 1
    while order <= s[0] and order == arr[order-1]:  
        order -= 1 # уменьшаем переменную порядка
    if order == 0: # если прошли по всему массиву и не нашли несоответствий
        prob = 0.0 # присваиваем минимальную вероятность
    
    for elem in rule:
        r = int(elem[0]) # длина сортировки
        p = float(elem[1]) # вероятность данной сортировки
        
        if order <= r: # если длина сортировки покрывает место найденного несоответствия
            prob = prob * (1-p) # считаем вероятность
    
    return 1-prob

try:
    t = int(input('Введите количество наборов входных данных: '))
    if not check_range(t, 1, 100): sys.exit('Выход за пределы диапазона! 1 <= t <= 100') 
    for i in range(t):
        s = list(map(int, input('Введите количество элементов и экспериментов: ').split()))
        if not check_range(s[0], 1, pow(10,5)) or not check_range(s[1], 1, pow(10,5)):
            sys.exit('Выход за пределы диапазона! 1 <= n,m <= 10^5')
        arr = list(map(int, input('Введите массив: ').split()))
        if not check_arr(arr, 1, s[0]): sys.exit('Выход за пределы диапазона! 1 <= a[i] <= n')
        if not check_equal(s[0], arr): sys.exit('Количество элементов массива не соответствует заданному номеру!')
        rule = []
        for i in range(s[1]):
            k = input('Введите правило: ').split()
            if not (k[0].isdigit() and check_range(int(k[0]), 1, s[0])): 
                sys.exit('В правиле должно быть целое число r: 1<=r<=n')
            if not check_float(float(k[1])):
                sys.exit('В правиле должно быть вещественное/целое число p: 0<=p<=1')
            rule.append(k) 
        print('Вероятность, что перестановка окажется отсортированной = {:6f}'.format(func(s, arr, rule)))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def count(n, p):
    medals = [0, 0, 0] # заполняем массив медалей нулями
    half = n//2-1 # ищем середину массива
    while p[half] == p[half+1]: # если средний элемент массива равен следующему
        half -= 1 # уменьшаем середину (медали которые будем распределять)
    
    # считаем золотые медали
    for i in range(1, half):
        if i*3 <= half+1-i:
            medals[0] += 1

    # считаем серебряные медали
    for i in range(medals[0], n):
        if (half+2-medals[0]-medals[1])//2 > medals[0]:
            medals[1] += 1
    
    # считаем бронзовые медали
    medals[2] = half+1-medals[0]-medals[1]

    return medals


try:
    t = int(input('Введите количество наборов входных данных: '))
    if not check_range(t, 1, 10000): sys.exit('Выход за пределы диапазона! 1 <= t <= 10000') 
    for i in range(t):
        n = int(input('Введите количество участников: '))
        if not check_range(n, 1, 4*pow(10,5)): sys.exit('Выход за пределы диапазона! 1 <= n <= 4*10^5')
        p = list(map(int, input('Введите массив: ').split()))
        if not check_arr(p, 0, pow(10,6)): sys.exit('Выход за пределы диапазона! 0 <= p[i] <= 10^6')
        if not check_equal(n, p): sys.exit('Количество элементов массива не соответствует заданному номеру!')
        if not check_decrease(p): sys.exit('Элементы массива должны быть в убывающем порядке!')
        # исключаем случай, когда медали нельзя будет распределить согласно условиям
        if n<6: print('Решения не существует: 0 0 0')
        else: print('Количество золотых, серебряных и бронзовых медалей = ', *count(n, p))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def good(a, n):
    diff = [] # заполняем массив модулями разностей соседних эл-ов
    for i in range(1, n): 
        diff.append(abs(a[i]-a[i-1]))
    
    # если массив состоит из 2 чисел и разница элементов = 1
    if len(diff) == 1 and diff[0] == 1: 
        return 1
    # если все элементы массива одинаковые
    elif diff.count(0) == len(diff): 
        return fact(n) 
    # если несколько раз встречается одинаковая разность
    elif contains(a, max(a)-min(a)):
        return rearr(n, a)
    else: return 0

def fact(n): # считаем факториал = длине массива
    res = 1
    for i in range(1, n+1):
        res = res*i
    return res

def rearr(n, a):
    res = 1
    a.sort(reverse=True) # сортируем массив по возрастанию
    frequency = {a.count(x):x for x in set(a)} # создаем словарь значениq и их количества
    temp = max(frequency.keys()) # выбираем максимальное
    for i in range(1, n):
        if a[i] == temp: # если элемент повторяется 
            res = res*temp # умножаем результат на сам элемент массива
        else: res = res*i # иначе считаем как обычный факториал
    return res

try:
    t = int(input('Введите количество наборов входных данных: '))
    if not check_range(t, 1, pow(10,4)): sys.exit('Выход за пределы диапазона! 1 <= t <= 10^4') 
    for i in range(t):
        n = int(input('Введите количество членов жюри: '))
        if not check_range(n, 2, 2*pow(10,5)): sys.exit('Выход за пределы диапазона! 2 <= n <= 2*10^5')
        a = list(map(int, input('Введите задачи: ').split()))
        if not check_arr(a, 1, pow(10,9)): sys.exit('Выход за пределы диапазона! 1 <= a[i] <= 10^9')
        if not check_equal(n, a): sys.exit('Количество элементов массива не соответствует заданному номеру!')
        m = 998244353
        print('Количество хороших перестановок = ', good(a, n)%m)
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

## Задача 8

Так как ответ может быть очень большим, выведите его по модулю 109+7 (то есть остаток при делении на 109+7). В случае если не существует ни одного подходящего массива (Аюб что-то перепутал) — выведите 0.

Входные данные\
Первая и единственная строка содержит три целых числа 𝑛, 𝑙 и 𝑟 (1≤𝑛≤2⋅105,1≤𝑙≤𝑟≤109) — размер массива и диапазон чисел в нём.

Выходные данные\

Примеры\

Примечание\
В первом примере возможны массивы [1,2],[2,1],[3,3].

Во втором примере единственный возможный массив — [2,2,2].

```python
def remake(t):
    rest = [0, 0, 0] # заполняем массив остатков нулями
    m = pow(10,9)+7 # задаем число для деления

    k = t[2]-t[1]+1 # считаем кол-во элементов в диапазоне
    for i in range(3): # заполняем массив остатков количеством
        rest[i] = k//3 # чисел с данным остатком от деления на 3
    if k%3 != 0:
        temp = t[1] % 3
        rest[temp] += 1

    cnt = [rest[0], rest[1], rest[2]] # заполняем массив-счетчик полученными значениями
    for i in range(2, t[0]+1): # считаем количество перестановок для остатков
        cnt0 = cnt[0]*rest[0] + cnt[1]*rest[2] + cnt[2]*rest[1]
        cnt1 = cnt[2]*rest[2] + cnt[0]*rest[1] + cnt[1]*rest[0]
        cnt2 = cnt[1]*rest[1] + cnt[0]*rest[2] + cnt[2]*rest[0]

        cnt[0] = cnt0 % m
        cnt[1] = cnt1 % m
        cnt[2] = cnt2 % m
    
    return cnt[0]

try:
    t = list(map(int, input('Введите размер массива и диапазон чисел: ').split()))
    if not check_range(t[0], 1, 2*pow(10,5)): sys.exit('Выход за пределы диапазона! 1 <= n <= 2*10^5') 
    if not check_range(t[1], 1, pow(10,9)): sys.exit('Выход за пределы диапазона! 1 <= l <= 10^9') 
    if not check_range(t[2], 1, pow(10,9)): sys.exit('Выход за пределы диапазона! 1 <= t <= 10^9') 
    print('Количество способов восстановить массив = ', remake(t))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

Выходные данные\

В первом тестовом примере мы можем выбрать 𝑑=−2.

Во втором тестовом примере мы можем выбрать 𝑑=−113.

В третьем тестовом примере мы не можем получить ни одного нуля в массиве 𝑐, какое бы значение 𝑑 мы ни выбрали.

```python
def zeros(n, a, b):
    temp = []
    for i in range(n): # заполняем массив частными != 0
        if a[i] != 0:
            temp.append(b[i]/a[i])

    if temp == []: # если массив пустой
        cnt = 0 # максимизировать нули не получится
    else:
        # создаем словарь частныхи и их количества
        frequency = {temp.count(x):x for x in set(temp)} 
        # выбираем самое частое повторение
        cnt = max(frequency.keys())

    return cnt

try:
    n = int(input('Введите количество элементов: '))
    if not check_range(n, 1, 2*pow(10,5)): sys.exit('Выход за пределы диапазона! 1 <= n <= 2*10^5')
    a = list(map(int, input('Введите массив: ').split()))
    if not check_arr(a, pow(-10,9), pow(10,9)): sys.exit('Выход за пределы диапазона! -10^9 <= a[i] <= 10^9')
    if not check_equal(n, a): sys.exit('Количество элементов массива не соответствует заданному номеру!')
    b = list(map(int, input('Введите массив: ').split()))
    if not check_arr(a, pow(-10,9), pow(10,9)): sys.exit('Выход за пределы диапазона! -10^9 <= b[i] <= 10^9')
    if not check_equal(n, a): sys.exit('Количество элементов массива не соответствует заданному номеру!')
    print('Максимальное количество нулей в массиве = ', zeros(n, a, b))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```

## Задача 10
Число называется сильным, если оно является степенью двойки или факториалом. Другими словами, число 𝑚 сильное, если существует неотрицательное целое число 𝑑 такое, что 𝑚=2𝑑 или 𝑚=𝑑!, где 𝑑!=1⋅2⋅…⋅𝑑 (в частности, 0!=1). Например, 1, 4 и 6 являются сильными, потому что 1=1!, 4=22, 6=3!, а 7, 10 или 18 не являются.

Вам дано положительное число 𝑛. Найдите минимальное число 𝑘 такое, что 𝑛 можно представить в виде суммы 𝑘 различных сильных чисел, или определите, что такого 𝑘 не существует.

Входные данные\
Во входных данных находятся несколько наборов входных данных. В первой строке находится одно целое число 𝑡 (1≤𝑡≤100) — количество наборов входных данных. Далее следуют наборы входных данных.

Описание одного набора входных данных содержит одно целое число 𝑛 (1≤𝑛≤1012).

Выходные данные\
Для каждого набора входных данных выведите ответ на отдельной строке.

Если 𝑛 не может быть представлено как сумма различных сильных чисел, выведите −1.

Иначе выведите одно положительное число — минимально возможное число 𝑘.

Пример\
входные данные\
4\
7\
11\
240\
17179869184\
выходные данные\
2\
3\
4\
1\
Примечание\
В первом примере 7 может быть представлена как 7=1+6, где 1 и 6 являются сильными. Так как 7 не является сильным числом, мы знаем, что минимальное значение 𝑘 в этом случае 𝑘=2.

Во втором примере один из способов разложить 11 в сумму трех сильных чисел такой: 11=1+4+6. Можно показать, что нельзя представить 11 в виде суммы двух или менее сильных чисел.

В третьем примере 240 может быть представлено как 240=24+32+64+120. Обратите внимание, что 240=120+120 не является корректным представлением, т. к. сильные числа должны быть различны.

В четвертом примере 17179869184=234, поэтому 17179869184 является сильным числом, и минимальное 𝑘 в этом случае 𝑘=1.

```python
def strong(n):
    if n%2 == 0: # если число четное
        # выбираем минимум из результатов выполнения функций
        res = min(degree(n), fact(n)) 
    else: # если нечетное
        # выбираем максимум из результатов выполнения функций
        res = max(degree(n), fact(n))
    return res if res > 0 else '-1' # если вариантов решения нет выводим -1

# проверяем является число степенью 2 и сколько элементов нужно для его представления
def degree(n):
    res, cnt = 1, 0
    while res < n:
        res *= 2
        cnt += 1
    # если является возвращаем 1 иначе количество эл-ов - 1
    return 1 if res == n else cnt-1

# проверяем является число факториалом и сколько элементов нужно для его представления
def fact(n):
    res, k, cnt = 1, 2, 0
    while res < n:
        res = res*k
        k += 1; cnt += 1
    # если является возвращаем 1 иначе количество эл-ов - 1
    return 1 if res == n else cnt-1

try:
    t = int(input('Введите количество наборов входных данных: '))
    if not check_range(t, 1, 100): sys.exit('Выход за пределы диапазона! 1 <= t <= 100') 
    for i in range(t):
        n = int(input('Введите число: '))
        if not check_range(n, 1, pow(10,12)): sys.exit('Выход за пределы диапазона! 1 <= n <= 10^12') 
        print('Минимально возможное число = ', strong(n))
except ValueError:
    print('Некорректные входные данные!')
except SystemExit as e:
    print('%s' % e.code) 
```
