```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—Ö–æ–∂–¥–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω —á–∏—Å–µ–ª
def check_range(n, a, b):
    if not a <= n <= b:
        return False
    return True

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –∑–∞–¥–∞–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é
def check_equal(n, arr):
    if not n == len(arr):
        return False
    return True

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö –ø—Ä–æ–ø–∏—Å–Ω—ã—Ö –∏ —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤
def check_ascii(arr):
    for i in range(len(arr)):
        if not (65<=ord(arr[i])<=90 or 97<=ord(arr[i])<=122):
            return False
    return True

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤ 'a' –∏ 'b'
def check_ascii_limited(arr):
    for i in range(len(arr)):
        if not 97<=ord(arr[i])<=98:
            return False
    return True

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Å—Å–∏–≤–∞ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ —á–∏—Å–µ–ª
def check_arr(arr, a, b):
    for elem in arr:
        if not a <= elem <= b:
            return False
    return True

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —á–∏—Å–ª–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –¥–∏–∞–ø–∞–∑–æ–Ω—É [0,1]
def check_float(num):
    if abs(1-num) <= 1:
        return True
    return False

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –≤ —É–±—ã–≤–∞—é—â–µ–º –ø–æ—Ä—è–¥–∫–µ
def check_decrease(arr):
    for i in range(len(arr)-1):
        if not arr[i] >= arr[i+1]:
            return False
    return True
```

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –º–∞—Å—Å–∏–≤–µ
def check(temp, el):
    for i in temp:
        if i == el:
            return 1

def flats(s):
    # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –º–∞—Å—Å–∏–≤–∞, –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ç–æ—á–∫–∏ –≤—Ö–æ–¥–∞ –∏ –≤—ã—Ö–æ–¥–∞
    temp, ent, out = [s[0]], 0, 0
    for i in range(len(s)-1):
        if s[i] == s[i+1] : # —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Å–æ—Å–µ–¥–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã
            if temp[0] == s[i+1]: # –µ—Å–ª–∏ –≤—Å—Ç—Ä–µ—á–∞–ª—Å—è —Ç–æ–ª—å–∫–æ –¥–∞–Ω–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç 
                ent = i+1 # —Å–º–µ—â–∞–µ–º —Ç–æ—á–∫—É –≤—Ö–æ–¥–∞
        elif check(temp, s[i+1]) != 1: # –ø—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —ç–ª–µ–º–µ–Ω—Ç –≤ –º–∞—Å—Å–∏–≤–µ
            out = i+1 # –µ—Å–ª–∏ –Ω–µ—Ç, —Ç–æ —Å–º–µ—â–∞–µ–º —Ç–æ—á–∫—É –≤—ã—Ö–æ–¥–∞
            temp.append(s[i+1]) # –∏ –¥–æ–±–∞–≤–ª—è–µ–º –≤ –º–∞—Å—Å–∏–≤

    return out-ent+1 

try:
    n = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–≤–∞—Ä—Ç–∏—Ä: ')) 
    if not check_range(n, 1, 100000): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n <= 100000')
    s = list(map(str, input('–í–≤–µ–¥–∏—Ç–µ –ø–æ–∫–µ–º–æ–Ω–æ–≤ –≤ –∫–≤–∞—Ä—Ç–∏—Ä–∞—Ö: ')))
    if not check_ascii(s): sys.exit('–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–æ–ª–∂–Ω—ã —Å–æ—Å—Ç–æ—è—Ç—å –∏–∑ —Å—Ç—Ä–æ—á–Ω—ã—Ö –∏ –ø—Ä–æ–ø–∏—Å–Ω—ã—Ö –±—É–∫–≤ –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞!')
    if not check_equal(n, s): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –ø–æ–∫–µ–º–æ–Ω–æ–≤ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–æ–º–µ—Ä—É –∫–≤–∞—Ä—Ç–∏—Ä!')
    else: print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∫–≤–∞—Ä—Ç–∏—Ä –¥–ª—è –ø–æ—Å–µ—â–µ–Ω–∏—è = ', flats(s))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code)
```

–ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—å –Ω–æ—Å–æ–∫ 6 –∏–∑ –ø—Ä–∞–≤–æ–≥–æ –≤ –ª–µ–≤—ã–π,\

–ø–µ—Ä–µ–∫—Ä–∞—Å–∏—Ç—å –Ω–æ—Å–æ–∫ 4 –≤ —Ü–≤–µ—Ç 1.\
–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –æ–Ω –ø–æ–ª—É—á–∏—Ç 3 –ø–∞—Ä—ã. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø–∞—Ä—ã (1,3), (2,4) –∏ (5,6).

```python
def change(one, two):
    while len(one) != len(two): # –ø–æ–∫–∞ –º–∞—Å—Å–∏–≤—ã –Ω–µ —Ä–∞–≤–Ω—ã 
        one.append(two[0]) # –¥–æ–±–∞–≤–ª—è–µ–º –≤ –º–µ–Ω—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç –∏–∑ –±–æ–ª—å—à–µ–≥–æ
        del two[0] # –∏ —É–¥–∞–ª—è–µ–º –µ–≥–æ –∏–∑ –±–æ–ª—å—à–µ–≥–æ
    return one, two

def check(one, two):
    for temp in one[:]: # –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ –æ–±—Ä–∞–∑–∞–º –º–∞—Å—Å–∏–≤–æ–≤
        for elem in two[:]:
            if temp == elem: # –µ—Å–ª–∏ –Ω–∞—Ö–æ–¥–∏–º –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ
                two.remove(elem) # —Ç–æ —É–¥–∞–ª—è–µ–º –∏–∑ –æ–±–æ–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤
                one.remove(temp)
                break
    return one, two  
    
def fill(s, c):
    # –∑–∞–Ω–æ—Å–∏–º —Ü–≤–µ—Ç–∞ –Ω–æ—Å–∫–æ–≤ –≤ '–ª–µ–≤—ã–π' –∏ '–ø—Ä–∞–≤—ã–π' –º–∞—Å—Å–∏–≤—ã
    count, j = 0, - 1
    l, r = [], []
    for i in range(int(s[1])):
        l.append(int(c[i]))
        j = i
    for i in range(int(s[2])):
        r.append(int(c[j+1])) 
        j += 1

    a, b = max(l, r), min(l, r) # –æ–ø—Ä–µ–¥–µ–ª—è–µ–º, –º–∞–∫—Å –∏ –º–∏–Ω –º–∞—Å—Å–∏–≤ 
    count = (len(a)-len(b))//2 # —Å—á–∏—Ç–∞–µ–º –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç
    change(b, a) # –¥–µ–ª–∞–µ–º '–ª–µ–≤—ã–π' –∏ '–ø—Ä–∞–≤—ã–π' –º–∞—Å—Å–∏–≤—ã —Ä–∞–≤–Ω—ã–º–∏
    check(a, b) # —É–±–∏—Ä–∞–µ–º –Ω–æ—Å–∫–∏ –æ–¥–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞ –≤ —Ä–∞–∑–Ω—ã—Ö –º–∞—Å—Å–∏–≤–∞—Ö
    
    return count + len(l) # –¥–æ–±–∞–≤–ª—è–µ–º –∫ —Å—á–µ—Ç—á–∏–∫—É –æ—Å—Ç–∞–≤—à—É—é—Å—è –¥–ª–∏–Ω—É –æ–¥–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞

try:
    t = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: "))
    if not check_range(t, 1, 1000): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= t <= 1000') 
    for i in range(t):
        s = list(map(int, input("–í–≤–µ–¥–∏—Ç–µ –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ—Å–∫–æ–≤, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–µ–≤—ã—Ö –∏ –ø—Ä–∞–≤—ã—Ö –Ω–æ—Å–∫–æ–≤: ").split()))
        if not check_range(s[0], 2, 2*pow(10,5)) or s[0]%2 != 0:
            sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! n - —á–µ—Ç–Ω–æ–µ —á–∏—Å–ª–æ –∏ 2 <= n <= 2*10^5')
        if not (check_range(s[1], 1, s[0]) or check_range(s[2], 1, s[0])) or s[1]+s[2]!=s[0]:
            sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 0 <= l,r <=n; l+r=n')
        c = list(map(int, input("–í–≤–µ–¥–∏—Ç–µ —Ü–≤–µ—Ç–∞ –Ω–æ—Å–∫–æ–≤: ").split()))
        if not check_arr(c, 1, s[0]): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= c[i] <= n')
        if not check_equal(s[0], c): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
        print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ–Ω–µ–≥, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –∑–∞–ø–ª–∞—Ç–∏—Ç—å = ', fill(s, c))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def same(n, s, t, a1, a2):
    difa, difb, pos = [], [], []
    for i in range(n):
        if s[i] != t[i]: # –ø—Ä–æ—Ö–æ–¥–∏–º—è –ø–æ —Å—Ç—Ä–æ–∫–∞–º
            if s[i] == 'a': 
                difa.append(i+1) # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –ø–æ–∑–∏—Ü–∏–∏ —Å –±—É–∫–≤–æ–π 'a'
            else:
                difb.append(i+1) # –∏ 'b'

    while len(difa) >= 2: # –ø–æ–∫–∞ –µ—Å—Ç—å >= 2 –ø–æ–∑–∏—Ü–∏–π 'a' –¥–ª—è –∏–∑–º–µ–Ω–µ–Ω–∏—è
        a1, a2 = difa.pop(0), difa.pop(0) # —É–±–∏—Ä–∞–µ–º –∏—Ö
        pos.append((a1, a2)) # –∏ –∑–∞–Ω–æ—Å–∏–º –≤ –º–∞—Å—Å–∏–≤ –ø–æ–∑–∏—Ü–∏–π
    
    while len(difb) >= 2: # –ø–æ–∫–∞ –µ—Å—Ç—å >= 2 –ø–æ–∑–∏—Ü–∏–π 'b' –¥–ª—è –∏–∑–º–µ–Ω–µ–Ω–∏—è
        a1, a2 = difb.pop(0), difb.pop(0) # —É–±–∏—Ä–∞–µ–º –∏—Ö
        pos.append((a1, a2)) # –∏ –∑–∞–Ω–æ—Å–∏–º –≤ –º–∞—Å—Å–∏–≤ –ø–æ–∑–∏—Ü–∏–π
    
    if len(difa) + len(difb) == 0: # –µ—Å–ª–∏ –º–∞—Å—Å–∏–≤—ã –ø—É—Å—Ç—ã
        return len(pos), pos # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç–≤–µ—Ç
        
    else: # –∏–Ω–∞—á–µ –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –ø–µ—Ä–≤—ã–µ –ø–æ–∑–∏—Ü–∏–∏ 'a' –∏–ª–∏ 'b'
        a1 = difa[0] 
        a2 = difb[0]
        pos.append((a1, a1)) # –∑–∞–ø–æ–ª–Ω—è–µ–º –∏–∑–º–µ–Ω–µ–Ω–∏—è –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ —Å–ª—É—á–∞—è
        pos.append((a1, a2))
    
        return len(pos), pos # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –æ—Ç–≤–µ—Ç

try:
    n = int(input("–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É —Å—Ç—Ä–æ–∫: "))
    if not check_range(n, 1, 2*pow(10,5)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n <= 2*10^5')
    s = list(map(str, input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É: ")))
    t = list(map(str, input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É: ")))
    if not check_equal(n, t) or not check_equal(n, s): 
        sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å—Ç—Ä–æ–∫–∏ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
    if not check_ascii_limited(s) or not check_ascii_limited(t):
        sys.exit('–°—Ç—Ä–æ–∫–∏ –¥–æ–ª–∂–Ω—ã —Å–æ—Å—Ç–æ—è—Ç—å –∏–∑ –ª–∞—Ç–∏–Ω—Å–∫–∏—Ö –±—É–∫–≤ ¬´a¬ª –∏ ¬´b¬ª')
    
    a1 = s.count('a') # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª-–≤–æ –±—É–∫–≤ 'a' –≤ –æ–±–µ–∏—Ö —Å—Ç—Ä–æ–∫–∞—Ö
    a2 = t.count('a')

    # –µ—Å–ª–∏ –Ω–µ—á–µ—Ç–Ω–æ–µ –∫–æ–ª-–≤–æ –±—É–∫–≤ 'a' –∏–ª–∏ 'b'
    if (a1+a2)%2 != 0 or (2*n-a1-a2)%2 != 0: print('–°—Ç—Ä–æ–∫–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏: -1')
    else:
        print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å = ', same(n, s, t, a1, a2)[0])
        print('–ü–æ–∑–∏—Ü–∏–∏ –¥–ª—è –æ–±–º–µ–Ω–∞ –±—É–∫–≤:')
        for elem in same(n, s, t, a1, a2)[1]: # –ø–æ—Å—Ç—Ä–æ—á–Ω–æ –≤—ã–≤–æ–¥–∏–º –ø–æ–∑–∏—Ü–∏–∏ –¥–ª—è –æ–±–º–µ–Ω–∞
                print(*elem)
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def func(s, arr, rule):
    prob = 1.0 # –∑–∞–¥–∞–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    order = s[0] # –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –ø–æ—Ä—è–¥–∫–∞ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    
    # –ø–æ–∫–∞ –ø–æ—Ä—è–¥–æ–∫ < –¥–ª–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ —ç–ª–µ–º–µ–Ω—Ç = —Å–≤–æ–µ–º—É –∏–Ω–¥–µ–∫—Å—É + 1
    while order <= s[0] and order == arr[order-1]:  
        order -= 1 # —É–º–µ–Ω—å—à–∞–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –ø–æ—Ä—è–¥–∫–∞
    if order == 0: # –µ—Å–ª–∏ –ø—Ä–æ—à–ª–∏ –ø–æ –≤—Å–µ–º—É –º–∞—Å—Å–∏–≤—É –∏ –Ω–µ –Ω–∞—à–ª–∏ –Ω–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–π
        prob = 0.0 # –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    
    for elem in rule:
        r = int(elem[0]) # –¥–ª–∏–Ω–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
        p = float(elem[1]) # –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
        
        if order <= r: # –µ—Å–ª–∏ –¥–ª–∏–Ω–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø–æ–∫—Ä—ã–≤–∞–µ—Ç –º–µ—Å—Ç–æ –Ω–∞–π–¥–µ–Ω–Ω–æ–≥–æ –Ω–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è
            prob = prob * (1-p) # —Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    
    return 1-prob

try:
    t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: '))
    if not check_range(t, 1, 100): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= t <= 100') 
    for i in range(t):
        s = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤: ').split()))
        if not check_range(s[0], 1, pow(10,5)) or not check_range(s[1], 1, pow(10,5)):
            sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n,m <= 10^5')
        arr = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤: ').split()))
        if not check_arr(arr, 1, s[0]): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= a[i] <= n')
        if not check_equal(s[0], arr): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
        rule = []
        for i in range(s[1]):
            k = input('–í–≤–µ–¥–∏—Ç–µ –ø—Ä–∞–≤–∏–ª–æ: ').split()
            if not (k[0].isdigit() and check_range(int(k[0]), 1, s[0])): 
                sys.exit('–í –ø—Ä–∞–≤–∏–ª–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ r: 1<=r<=n')
            if not check_float(float(k[1])):
                sys.exit('–í –ø—Ä–∞–≤–∏–ª–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ/—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ p: 0<=p<=1')
            rule.append(k) 
        print('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –æ–∫–∞–∂–µ—Ç—Å—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π = {:6f}'.format(func(s, arr, rule)))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

–≠—Ç–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ —è–≤–ª—è–µ—Ç—Å—è —Ö–æ—Ä–æ—à–µ–π.

```python
def good(a, n):
    diff = [] # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ –º–æ–¥—É–ª—è–º–∏ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π —Å–æ—Å–µ–¥–Ω–∏—Ö —ç–ª-–æ–≤
    for i in range(1, n): 
        diff.append(abs(a[i]-a[i-1]))
    
    # –µ—Å–ª–∏ –º–∞—Å—Å–∏–≤ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ 2 —á–∏—Å–µ–ª –∏ —Ä–∞–∑–Ω–∏—Ü–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ = 1
    if len(diff) == 1 and diff[0] == 1: 
        return 1
    # –µ—Å–ª–∏ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ
    elif diff.count(0) == len(diff): 
        return fact(n) 
    # –µ—Å–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –æ–¥–∏–Ω–∞–∫–æ–≤–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å
    elif contains(a, max(a)-min(a)):
        return rearr(n, a)
    else: return 0

def fact(n): # —Å—á–∏—Ç–∞–µ–º —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª = –¥–ª–∏–Ω–µ –º–∞—Å—Å–∏–≤–∞
    res = 1
    for i in range(1, n+1):
        res = res*i
    return res

def rearr(n, a):
    res = 1
    a.sort(reverse=True) # —Å–æ—Ä—Ç–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é
    frequency = {a.count(x):x for x in set(a)} # —Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –∑–Ω–∞—á–µ–Ω–∏q –∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
    temp = max(frequency.keys()) # –≤—ã–±–∏—Ä–∞–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ
    for i in range(1, n):
        if a[i] == temp: # –µ—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è 
            res = res*temp # —É–º–Ω–æ–∂–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ —Å–∞–º —ç–ª–µ–º–µ–Ω—Ç –º–∞—Å—Å–∏–≤–∞
        else: res = res*i # –∏–Ω–∞—á–µ —Å—á–∏—Ç–∞–µ–º –∫–∞–∫ –æ–±—ã—á–Ω—ã–π —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª
    return res

try:
    t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: '))
    if not check_range(t, 1, pow(10,4)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= t <= 10^4') 
    for i in range(t):
        n = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–ª–µ–Ω–æ–≤ –∂—é—Ä–∏: '))
        if not check_range(n, 2, 2*pow(10,5)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 2 <= n <= 2*10^5')
        a = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ –∑–∞–¥–∞—á–∏: ').split()))
        if not check_arr(a, 1, pow(10,9)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= a[i] <= 10^9')
        if not check_equal(n, a): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
        m = 998244353
        print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ö–æ—Ä–æ—à–∏—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ = ', good(a, n)%m)
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def remake(t):
    rest = [0, 0, 0] # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ –æ—Å—Ç–∞—Ç–∫–æ–≤ –Ω—É–ª—è–º–∏
    m = pow(10,9)+7 # –∑–∞–¥–∞–µ–º —á–∏—Å–ª–æ –¥–ª—è –¥–µ–ª–µ–Ω–∏—è

    k = t[2]-t[1]+1 # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ
    for i in range(3): # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ –æ—Å—Ç–∞—Ç–∫–æ–≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º
        rest[i] = k//3 # —á–∏—Å–µ–ª —Å –¥–∞–Ω–Ω—ã–º –æ—Å—Ç–∞—Ç–∫–æ–º –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 3
    if k%3 != 0:
        temp = t[1] % 3
        rest[temp] += 1

    cnt = [rest[0], rest[1], rest[2]] # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤-—Å—á–µ—Ç—á–∏–∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
    for i in range(2, t[0]+1): # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –¥–ª—è –æ—Å—Ç–∞—Ç–∫–æ–≤
        cnt0 = cnt[0]*rest[0] + cnt[1]*rest[2] + cnt[2]*rest[1]
        cnt1 = cnt[2]*rest[2] + cnt[0]*rest[1] + cnt[1]*rest[0]
        cnt2 = cnt[1]*rest[1] + cnt[0]*rest[2] + cnt[2]*rest[0]

        cnt[0] = cnt0 % m
        cnt[1] = cnt1 % m
        cnt[2] = cnt2 % m
    
    return cnt[0]

try:
    t = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ —Ä–∞–∑–º–µ—Ä –º–∞—Å—Å–∏–≤–∞ –∏ –¥–∏–∞–ø–∞–∑–æ–Ω —á–∏—Å–µ–ª: ').split()))
    if not check_range(t[0], 1, 2*pow(10,5)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n <= 2*10^5') 
    if not check_range(t[1], 1, pow(10,9)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= l <= 10^9') 
    if not check_range(t[2], 1, pow(10,9)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= t <= 10^9') 
    print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –º–∞—Å—Å–∏–≤ = ', remake(t))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

```python
def zeros(n, a, b):
    temp = []
    for i in range(n): # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ —á–∞—Å—Ç–Ω—ã–º–∏ != 0
        if a[i] != 0:
            temp.append(b[i]/a[i])

    if temp == []: # –µ—Å–ª–∏ –º–∞—Å—Å–∏–≤ –ø—É—Å—Ç–æ–π
        cnt = 0 # –º–∞–∫—Å–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –Ω—É–ª–∏ –Ω–µ –ø–æ–ª—É—á–∏—Ç—Å—è
    else:
        # —Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —á–∞—Å—Ç–Ω—ã—Ö–∏ –∏ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞
        frequency = {temp.count(x):x for x in set(temp)} 
        # –≤—ã–±–∏—Ä–∞–µ–º —Å–∞–º–æ–µ —á–∞—Å—Ç–æ–µ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ
        cnt = max(frequency.keys())

    return cnt

try:
    n = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤: '))
    if not check_range(n, 1, 2*pow(10,5)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n <= 2*10^5')
    a = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤: ').split()))
    if not check_arr(a, pow(-10,9), pow(10,9)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! -10^9 <= a[i] <= 10^9')
    if not check_equal(n, a): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
    b = list(map(int, input('–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤: ').split()))
    if not check_arr(a, pow(-10,9), pow(10,9)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! -10^9 <= b[i] <= 10^9')
    if not check_equal(n, a): sys.exit('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞–¥–∞–Ω–Ω–æ–º—É –Ω–æ–º–µ—Ä—É!')
    print('–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–π –≤ –º–∞—Å—Å–∏–≤–µ = ', zeros(n, a, b))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```

–û–ø–∏—Å–∞–Ω–∏–µ –æ–¥–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (1‚â§ùëõ‚â§1012).

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–¥–∏–Ω –∏–∑ —Å–ø–æ—Å–æ–±–æ–≤ —Ä–∞–∑–ª–æ–∂–∏—Ç—å 11 –≤ —Å—É–º–º—É —Ç—Ä–µ—Ö —Å–∏–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª —Ç–∞–∫–æ–π: 11=1+4+6. –ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –Ω–µ–ª—å–∑—è –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å 11 –≤ –≤–∏–¥–µ —Å—É–º–º—ã –¥–≤—É—Ö –∏–ª–∏ –º–µ–Ω–µ–µ —Å–∏–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª.

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ 240 –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ –∫–∞–∫ 240=24+32+64+120. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ 240=120+120 –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ–º, —Ç. –∫. —Å–∏–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã.

–í —á–µ—Ç–≤–µ—Ä—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ 17179869184=234, –ø–æ—ç—Ç–æ–º—É 17179869184 —è–≤–ª—è–µ—Ç—Å—è —Å–∏–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º, –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ ùëò –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ ùëò=1.

```python
def strong(n):
    if n%2 == 0: # –µ—Å–ª–∏ —á–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ
        # –≤—ã–±–∏—Ä–∞–µ–º –º–∏–Ω–∏–º—É–º –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π
        res = min(degree(n), fact(n)) 
    else: # –µ—Å–ª–∏ –Ω–µ—á–µ—Ç–Ω–æ–µ
        # –≤—ã–±–∏—Ä–∞–µ–º –º–∞–∫—Å–∏–º—É–º –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π
        res = max(degree(n), fact(n))
    return res if res > 0 else '-1' # –µ—Å–ª–∏ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ —Ä–µ—à–µ–Ω–∏—è –Ω–µ—Ç –≤—ã–≤–æ–¥–∏–º -1

# –ø—Ä–æ–≤–µ—Ä—è–µ–º —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ —Å—Ç–µ–ø–µ–Ω—å—é 2 –∏ —Å–∫–æ–ª—å–∫–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω—É–∂–Ω–æ –¥–ª—è –µ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
def degree(n):
    res, cnt = 1, 0
    while res < n:
        res *= 2
        cnt += 1
    # –µ—Å–ª–∏ —è–≤–ª—è–µ—Ç—Å—è –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 1 –∏–Ω–∞—á–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª-–æ–≤ - 1
    return 1 if res == n else cnt-1

# –ø—Ä–æ–≤–µ—Ä—è–µ–º —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–æ–º –∏ —Å–∫–æ–ª—å–∫–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω—É–∂–Ω–æ –¥–ª—è –µ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
def fact(n):
    res, k, cnt = 1, 2, 0
    while res < n:
        res = res*k
        k += 1; cnt += 1
    # –µ—Å–ª–∏ —è–≤–ª—è–µ—Ç—Å—è –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 1 –∏–Ω–∞—á–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª-–æ–≤ - 1
    return 1 if res == n else cnt-1

try:
    t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: '))
    if not check_range(t, 1, 100): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= t <= 100') 
    for i in range(t):
        n = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
        if not check_range(n, 1, pow(10,12)): sys.exit('–í—ã—Ö–æ–¥ –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞! 1 <= n <= 10^12') 
        print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ —á–∏—Å–ª–æ = ', strong(n))
except ValueError:
    print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ!')
except SystemExit as e:
    print('%s' % e.code) 
```
